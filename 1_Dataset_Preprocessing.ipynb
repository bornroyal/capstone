## Creating Foundation column 

```python
# Creating a new column called foundation that will be a combination of other columns
df['Foundation'] = df['brand'] + ' ' +  df['product'] + ' ' + df['specific']
```

```python
df.head()
```

```python
# Creating a function to remove the # sign in the hex values to convert to RGB
def fix_hex(x):
    x= x.strip('#')
    return x
```

```python
# using our # hex removal function on the hex column
df['hex'] = df['hex'].apply(fix_hex)
```

```python
df.head()
```

```python
# Setting the foundation as the index
df.set_index(df['Foundation'], inplace=True)
```

```python
# converting the column titled 'name' to shade 
df.rename(columns={'name':'shade'}, inplace=True)
```

```python
df.head()
```

```python
# Dropping columns not useful for our analysis
df.drop(columns=['brand', 'product', 'description', 'imgSrc', 'imgAlt', 'specific', 'colorspace', 'Foundation'], inplace=True)
```

```python
df.head()
```

# Converting HEX values to RGB

```python
# Using a lambda function to turn the HEX into a tupe of hex values in RGB space
df['hex'] = df['hex'].apply(lambda x: tuple(hex(x).rgb))
```

```python
df.head()
```

# Creating RGB columns

```python
# Now we will create 3 new columns of red green blue to store our values
red =[]
green = []
blue = []

for i in df['hex']: # Lopping through every hex value in the dataframe 
    red.append(i[0])
    green.append(i[1])
    blue.append(i[2])
```

```python
df['red'] = red
df['green'] = green
df['blue'] = blue
```
